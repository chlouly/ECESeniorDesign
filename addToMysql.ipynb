```python
def add_data_to_database(data):

    # SQL query to insert data into the table
    insert_query = """
    INSERT INTO mytable (passage, question, choice_A, choice_B, choice_C, choice_D, correct_answer, rationale, difficulty, domain, id_collegeboard) 
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    # data = (1, "Sample passage", "Sample question", "Choice A", "Choice B", "Choice C", "Choice D", "A", "Sample rationale", "Easy", "Sample domain")


    cursor.execute(insert_query, data)
    connection.commit()

    print("Data added successfully!")
```
