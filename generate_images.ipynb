```python
from openai import OpenAI
import os
from PIL import Image
import requests
from io import BytesIO
seed = 12345
# Set your OpenAI API key here
# openai.api_key = 
client = OpenAI(api_key='your key')
def generate_image(prompt):
    try:


        response = client.images.generate(
            model="dall-e-3",
            prompt=prompt,
            n=1,
            size="1024x1024",
            quality="standard")
        print(response)
        image_url = response.data[0].url
        print(image_url)
        return image_url
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

def download_image(url, file_path):
    response = requests.get(url)
    image = Image.open(BytesIO(response.content))
    image.save(file_path)




```

```python

directory = "pokidips_monsters"
os.makedirs(directory, exist_ok=True)

for i in range(50):
    monster_prompt = f"Using the character from gen_id A4nU8DlJC3FuBfyL, create a colorful, cartoonish 2D creature, similar to a Pokémon, with vibrant colors monster. The creature should look friendly or mystical and have a white background, suitable for a video game. Don't include any text or logos in the image! I want to have only the creature in the image!"
    print(f"Generating image {i+1}/50")
    image_url = generate_image(monster_prompt)
    if image_url:
        file_path = os.path.join(directory, f"{i+1+13}.png")
        download_image(image_url, file_path)
        print(f"Saved: {file_path}")
    else:
        print(f"Failed to generate image {i+1}")
```
